```python
import datetime
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
start = "2020-01-01"
end = "2021-10-12"
jsw = yf.download("JSWSTEEL.BO",start,end)
sail = yf.download("SAIL.BO",start,end)
tata = yf.download("TATASTEEL.BO",start,end)
jsw["Open"].plot(label = "JSWSTEEL.BO", figsize = (15,7))
sail["Open"].plot(label = "SAIL.BO")
tata["Open"].plot(label = "TATASTEEL.BO")
plt.legend()
plt.title("stock prices of jsw,sail,tata")
```

```python
# VOLUME TRADED
jsw["Volume"].plot(label = "JSWSTEEL.BO",figsize=(15,7))
sail["Volume"].plot(label = "SAIL.BO")
tata["Volume"].plot(label = "TATASTEEL.BO")
plt.title("Volume of Stock traded")
plt.legend()
```

```python
# MARKET CAP
jsw["MarktCap"] = jsw["Open"]*jsw["Volume"]
sail["MarktCap"] = sail["Open"]*sail["Volume"]
tata["MarktCap"] = tata["Open"]*tata["Volume"]
jsw["MarktCap"].plot(label = "JSWSTEEL.BO",figsize=(15,7))
sail["MarktCap"].plot(label = "SAIL.BO")
tata["MarktCap"].plot(label = "TATASTEEL.BO")
plt.title("Market Cap")
plt.legend()
```

```python
# MOVING AVERAGES
# MA of JSW STEEL
jsw["MA50"] = jsw["Open"].rolling(50).mean()
jsw["MA200"] = jsw["Open"].rolling(200).mean()
jsw["Open"].plot(label = "Open",figsize = (15,7))
jsw["MA50"].plot(label = "50 DAYS MA")
jsw["MA200"].plot(label = "200 DAYS MA")
plt.legend()
```

```python
# MA of SAIL
sail["MA50"] = sail["Open"].rolling(50).mean()
sail["MA200"] = sail["Open"].rolling(200).mean()
sail["Open"].plot(label = "Open",figsize = (15,7))
sail["MA50"].plot(label = "50 DAYS MA")
sail["MA200"].plot(label = "200 DAYS MA")
plt.legend()
```

```python
# MA of TATA STEEL
tata["MA50"] = tata["Open"].rolling(50).mean()
tata["MA200"] = tata["Open"].rolling(200).mean()
tata["Open"].plot(label = "Open",figsize = (15,7))
tata["MA50"].plot(label = "50 DAYS MA")
tata["MA200"].plot(label = "200 DAYS MA")
plt.legend()
```

```python
# SCATTER PLOT MATRIX
import pandas as pd
data = pd.concat([jsw["Open"],sail["Open"],tata["Open"]], axis=1)
data.columns = ["jswOpen","sailOpen","tataOpen"]
scatter_matrix(data, figsize=(8,8))
```

```python
# VOLATILITY
jsw["returns"]= (jsw["Close"]/jsw["Close"].shift(1))-1
sail["returns"]= (sail["Close"]/sail["Close"].shift(1))-1
tata["returns"]= (tata["Close"]/tata["Close"].shift(1))-1
jsw["returns"].hist(bins = 100,label="JSWSTEEL.BO",alpha = 0.5,figsize = (15,7))
sail["returns"].hist(bins = 100, label = "SAIL.BO",alpha = 0.5)
tata["returns"].hist(bins = 100, label = "TATASTEEL.BO",alpha = 0.5)
plt.legend()
```
