```python
import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas.plotting import scatter_matrix
import yfinance as yf
start = "2021-06-01"
end = "2021-10-25"
tata_motors = yf.download("TATAMOTORS.BO",start,end)
maruti = yf.download("MARUTI.BO",start,end)
ashok_leyland = yf.download("ASHOKLEY.BO",start,end)
tata_motors["Open"].plot(label = "tata motors", figsize = (15,7))
maruti["Open"].plot(label = "maruti")
ashok_leyland["Open"].plot(label = "ashok leyland")
plt.legend()
plt.title("stock prices of tata motors,maruti,ashok leyland")


# tata motors price is completely higher than other two


```

```python
tata_motors["Volume"].plot(label = "tata motors",figsize=(15,7))
maruti["Volume"].plot(label = "maruti")
ashok_leyland["Volume"].plot(label = "ashok leyland")
plt.title("Volume of Stock traded")
plt.legend()

# tata motors is traded high in market other than two
```

```python
tata_motors["MarktCap"] = tata_motors["Open"]*tata_motors["Volume"]
maruti["MarktCap"] = maruti["Open"]*maruti["Volume"]
ashok_leyland["MarktCap"] = ashok_leyland["Open"]*ashok_leyland["Volume"]
tata_motors["MarktCap"].plot(label = "tata motors",figsize=(15,7))
maruti["MarktCap"].plot(label = "maruti")
ashok_leyland["MarktCap"].plot(label = "ashok leyland")
plt.title("Market Cap")
plt.legend()

# tata marketcap is higher than two
```

```python
tata_motors["MA9"] = tata_motors["Open"].rolling(9).mean()
tata_motors["MA50"] = tata_motors["Open"].rolling(50).mean()
tata_motors["Open"].plot(label = "Open",figsize = (15,7))
tata_motors["MA9"].plot(label = "9 DAYS MA")
tata_motors["MA50"].plot(label = "50 DAYS MA")

plt.legend()
```

```python
maruti["MA9"] = maruti["Open"].rolling(9).mean()
maruti["MA50"] = maruti["Open"].rolling(50).mean()
maruti["Open"].plot(label = "Open",figsize = (15,7))
maruti["MA9"].plot(label = "9 DAYS MA")
maruti["MA50"].plot(label = "50 DAYS MA")

plt.legend()
```

```python
ashok_leyland["MA9"] = ashok_leyland["Open"].rolling(9).mean()
ashok_leyland["MA50"] = ashok_leyland["Open"].rolling(50).mean()
ashok_leyland["Open"].plot(label = "Open",figsize = (15,7))
ashok_leyland["MA9"].plot(label = "9 DAYS MA")
ashok_leyland["MA50"].plot(label = "50 DAYS MA")

plt.legend()
```

```python
data = pd.concat([tata_motors["Open"],maruti["Open"],ashok_leyland["Open"]], axis=1)
data.columns = ["tata_motorsOpen","marutiOpen","ashok_leylandOpen"]
scatter_matrix(data, figsize=(8,8))
```

```python
tata_motors["returns"]= (tata_motors["Close"]/tata_motors["Close"].shift(1))-1
maruti["returns"]= (maruti["Close"]/maruti["Close"].shift(1))-1
ashok_leyland["returns"]= (ashok_leyland["Close"]/ashok_leyland["Close"].shift(1))-1
tata_motors["returns"].hist(bins = 100,label="tata motors",alpha = 0.5,figsize = (15,7))
maruti["returns"].hist(bins = 100, label = "maruti",alpha = 0.5)
ashok_leyland["returns"].hist(bins = 100, label = "ashok leyland",alpha = 0.5)
plt.legend()
# tata motors is best in automobile because it has high volatility,volume 
# moving average also fine
```
